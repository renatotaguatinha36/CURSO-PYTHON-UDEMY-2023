```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
```

```python
df.drop(columns=['ID'], inplace=True)
df.head(3)
```

```python
plt.pie(df['Satisfaction'].value_counts(), labels=['Neutral or Dissatisfied', 'Satisfied'], autopct='%1.1f%%')
plt.show()
```

```python
cols = ['Gender', 'Customer Type', 'Type of Travel', 'Class', 'Satisfaction']
plt.figure(figsize=(15, 15))
for i, col in enumerate(cols):
    plt.subplot(3, 2, i + 1)
    sns.countplot(x=col, data=df)
plt.show()
```

```python
df.hist(bins=20, figsize=(20, 20), color='green')
plt.show()
```

```python
df['Satisfaction'].unique()
```

```python
label_encoder = preprocessing.LabelEncoder()

object_columns = df.select_dtypes(include=['object']).drop(columns=['Satisfaction']).columns

for column in object_columns:
    df[column] = label_encoder.fit_transform(df[column])

df.head()
```
